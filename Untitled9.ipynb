```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure necessary NLTK resources are downloaded
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load your dataset
data = pd.read_csv('/content/youtube.csv')

# Function to preprocess text
def preprocess_text(text):
    tokens = word_tokenize(text.lower())  # Tokenize and convert to lower case
    tokens = [t for t in tokens if t.isalpha()]  # Keep only alphabetic tokens
    tokens = [t for t in tokens if t not in stopwords.words('english')]  # Remove stopwords
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(t) for t in tokens]  # Lemmatize words
    return ' '.join(tokens)

# Apply preprocessing to the title and description columns
data['processed_title'] = data['title'].apply(preprocess_text)
data['processed_description'] = data['description'].apply(preprocess_text)

# Function to get sentiment
def get_sentiment(text):
    sentiment = TextBlob(text)
    return sentiment.polarity, sentiment.subjectivity

# Apply the function to the processed text columns
data['title_polarity'], data['title_subjectivity'] = zip(*data['processed_title'].apply(get_sentiment))
data['description_polarity'], data['description_subjectivity'] = zip(*data['processed_description'].apply(get_sentiment))

# Visualization of title sentiment polarity
plt.figure(figsize=(10, 6))
sns.histplot(data['title_polarity'], bins=30, kde=True)
plt.title('Distribution of Title Sentiment Polarity')
plt.xlabel('Polarity')
plt.ylabel('Frequency')
plt.show()

# Visualization of description sentiment polarity
plt.figure(figsize=(10, 6))
sns.histplot(data['description_polarity'], bins=30, kde=True)
plt.title('Distribution of Description Sentiment Polarity')
plt.xlabel('Polarity')
plt.ylabel('Frequency')
plt.show()

# Additional visualizations can be added for subjectivity or other insights as needed.
```
